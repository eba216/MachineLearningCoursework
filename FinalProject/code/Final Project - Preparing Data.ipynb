This describes 7 posibilities for classifying a case of COVID-19,

Here we translate the text. 

Confirmed by association applies when the case reported being a contact with a positive COVID-19 (and this is registered in SISVER) and:
The case was not sampled or the sample was invalid.

Confirmed by ruling only applies to deaths under the following conditions:
The case was not sampled or a sample was taken, but the sample was invalid.

Confirmed applies when:
The case has a laboratory sample or antigenic test and was positive for SARS-CoV-2, regardless of whether the case has a clinical epidemiological association.

Invalid applies when the case does not have a clinical epidemiological association, nor a COVID-19 ruling. A laboratory sample was taken and it was invalid.

Not carried out applies when the case does not have a clinical epidemiological association, or a COVID-19 ruling and a laboratory sample was taken and it was not processed.

Suspect applies when:
The case does not have a clinical epidemiological association, nor is it dictated by COVID-19 and no sample was taken, or a laboratory sample was taken and the result is pending, regardless of another condition.

Negative applies when the case:
1. A laboratory sample was taken and it was: negative for SARS-COV-2 or positive for any other respiratory virus (Influenza, RSV, Bocavirus, others) regardless of whether this case has an epidemiological clinical association or a COVID-19 ruling .
2. An antigenic sample was taken that was negative for SARS-COV-2 and the case was not taken from a laboratory sample or confirmed by epidemiological association or by clinical epidemiological opinion.

Now we rename and translate the columns.  

```python
translations = {
    "SEXO": "Sex", 
    "EDAD":"Age",
    "EMBARAZO":"Pregnant",
    "NEUMONIA":"Pneumonia",
    "DIABETES":"Diabetes", 
    "EPOC":"COPD", 
    "ASMA":"Athsma",
    "INMUSUPR":"Immunosuppressed", 
    "HIPERTENSION":"Hypertension", 
    "OTRA_COM":"Other Conds.", 
    "CARDIOVASCULAR":"Cardiovascular",
    "OBESIDAD":"Obesity",
    "RENAL_CRONICA":"Chronic Kidney Disease", 
    "TABAQUISMO":"Smoking",
    "UCI":"ICU"
}
raw_data_es = raw_data_es.rename(columns = translations)
raw_data_es
```

We now prepare to convert our most of our data values from 1 and 2 to 1 and 0, representing yes and no (or female and male).

We note that a value of 97 means not applicable, which we interpret as a 0 (no).
We also note that a value of 98 means to ignore, which we interpret as a 0 (no) for ICU and as NaN otherwise.
We recall that a value of 99 means not specifified, which we interpret as a Nan .

```python
#ICU is a rare event
count = raw_data_es.groupby(["ICU"]).count()
count
```

```python
def re_number(x:int):
    if x ==1.0:
        return 1
    elif x == 99 or x == 98:
        return 
    else:
        return 0
    
    
def re_number_ICU(x:int):
    if x == 1.0:
        return 1
    elif x == 99:
        return 0
    else:
        return 0

# loop to clean up data
for i in range(0,15):
    # skip Age
    if i == 1:
        continue
    raw_data_es.iloc[:,i] = raw_data_es.iloc[:,i].apply(re_number)
    # second application is to clean up data, for some reason we
    # have a mixture of floats and ints otherwise
    raw_data_es.iloc[:,i] = raw_data_es.iloc[:,i].apply(re_number)

raw_data_es["ICU"] = raw_data_es["ICU"].apply(re_number_ICU)

raw_data_es = raw_data_es.dropna(axis=0)
raw_data_es = raw_data_es.reset_index(drop=True)
raw_data_es
```

Finally, we export the data into our processed data folder.
